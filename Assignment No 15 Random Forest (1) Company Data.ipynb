```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Company_Data.csv")
df.head()
```

```python
#Creating dummy vairables for ['Undergrad','Marital.Status','Urban'] dropping first dummy variable
df=pd.get_dummies(df,columns=['ShelveLoc','Urban','US'], drop_first=True)
df.head()
```

```python
#Creating new cols TaxInc and dividing 'Taxable.Income' cols on the basis of [10002,30000,99620] for Risky and Good
df["sales"] = pd.cut(df["Sales"], bins = [0,10,100], labels = ["Medium","Good"])
df["sales"]
```

```python
#After creation of new col. TaxInc also made its dummies var concating right side of df
df = pd.get_dummies(df,columns = ["sales"],drop_first=True)
df
```

```python
df.tail()
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(df.iloc[:,1:])
df_norm.tail(10)
```

```python
# Declaring features & target
X = df_norm.drop(['sales_Good'], axis=1)
y = df_norm['sales_Good']
```

```python
y.tail()
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting data into train & test
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Choosing RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Creating new model testing with new parameters
forest = RandomForestClassifier(n_estimators=100,max_depth=10,min_samples_split=400,criterion='gini')  # n_estimators is the number of decision trees
forest.fit(Xtrain, ytrain)
```

```python
print('Train accuracy: {}'.format(forest.score(Xtrain, ytrain)))
print('Test accuracy: {}'.format(forest.score(Xtest, ytest)))
```

```python
# Creating new model testing with new parameters
forest_new = RandomForestClassifier(n_estimators=100,max_depth=10,min_samples_split=20,criterion='gini')  # n_estimators is the number of decision trees
forest_new.fit(Xtrain, ytrain)
```

```python
print('Train accuracy: {}'.format(forest_new.score(Xtrain, ytrain)))
print('Test accuracy: {}'.format(forest_new.score(Xtest, ytest)))
```
